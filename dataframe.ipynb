```python
import pandas as pd
```

```python
#data frame = 2D
data={
    'name':['ram','hari','shyam'],
    'age':[20,25,30],
    'city':['kathmandu','lalitpur','bhaktapur']
}

df=pd.DataFrame(data,columns=['name','age','city'])
print(df)
#saving to csv
df.to_csv('output.csv', index=False)#removes index
```
