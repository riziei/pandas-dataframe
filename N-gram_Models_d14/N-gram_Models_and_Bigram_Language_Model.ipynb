# N-Grams in NLP

## What are N-grams?

N-grams are continuous sequences of **N words** from a text or sentence. They are used in NLP to understand word patterns and relationships.

---

## Types of N-grams

### 1. Unigram (N = 1)
Single words from a sentence.

**Example:**
Sentence: "I love NLP"

Unigrams:
- I
- love
- NLP

---

### 2. Bigram (N = 2)
Pairs of consecutive words.

**Example:**
Sentence: "I love NLP"

Bigrams:
- I love
- love NLP

---

### 3. Trigram (N = 3)
Sequence of three consecutive words.

**Example:**
Sentence: "I love NLP"

Trigrams:
- I love NLP

---

## Why N-grams are important

- Capture word relationships
- Help in text prediction
- Used in language models
- Power autocomplete systems
- Used in machine translation and speech recognition

---

N-grams are the foundation of statistical NLP models that help machines understand and predict language using word sequences.

Import Libraries

````python
import re
import random
from collections import Counter, defaultdict
import pandas as pd
````

Define Training Corpus

We need some text data to learn word relationships.

````python
corpus = """
I love natural language processing.
I love machine learning.
Machine learning is fun.
Natural language processing is interesting.
"""

print(corpus)
````

Text Preprocessing

Convert text to lowercase and tokenize into words.

````python
# Convert to lowercase
corpus = corpus.lower()

# Remove punctuation
corpus = re.sub(r"[^\w\s]", "", corpus)

# Split into words
tokens = corpus.split()

print("Tokens:")
print(tokens)
````

Generate Unigrams

````python
# Unigrams are individual words
unigrams = tokens

print("Unigrams:")
print(unigrams)
````

Generate Bigrams

````python
# Create word pairs

bigrams = []

for i in range(len(tokens) - 1):
    bigrams.append((tokens[i], tokens[i + 1]))

print("Bigrams:")
print(bigrams)
````

Generate Trigrams

````python
# Create sequences of 3 words

trigrams = []

for i in range(len(tokens) - 2):
    trigrams.append((tokens[i], tokens[i + 1], tokens[i + 2]))

print("Trigrams:")
print(trigrams)
````

Count Bigram Frequencies

Count how many times each word pair appears.

````python
bigram_counts = Counter(bigrams)

print("Bigram Frequencies:\n")

for bigram, count in bigram_counts.items():
    print(f"{bigram}: {count}")
````

````python
df = pd.DataFrame(bigram_counts.items(), columns=["Bigram", "Count"])
df.sort_values(by="Count", ascending=False)
````

Count Individual Word Frequencies

````python
word_counts = Counter(tokens)

print("Word Frequencies:\n")

for word, count in word_counts.items():
    print(f"{word}: {count}")
````

## Bigram Language Model

A Bigram Model predicts a word based on the previous word.

Example:

Sentence:
"I love NLP"

Bigrams:
- (I, love)
- (love, NLP)

The probability of a word depends only on the word before it.

---

##  Bigram Probability Calculation

A Bigram Language Model predicts the next word based on the current word.

The probability of seeing a word **w₂** after a word **w₁** is calculated as:

$$
P(w_2|w_1)
=
\frac{Count(w_1,w_2)}
{Count(w_1)}
$$

### Where:

- **Count(w₁, w₂)** = Number of times the bigram `(w₁, w₂)` appears in the corpus
- **Count(w₁)** = Number of times the word `w₁` appears in the corpus
---

### Example

Suppose our corpus contains:

```text
i love nlp
i love machine learning
```

Bigrams:

```text
(i, love)
(love, nlp)
(i, love)
(love, machine)
(machine, learning)
```

Counts:

```text
Count(i, love) = 2
Count(i) = 2
```

Therefore:

$$
P(\text{love} \mid \text{i})
=
\frac{\text{Count}(i,\ love)}
{\text{Count}(i)}
=
\frac{2}{2}
=
1.0
$$

This means that whenever the word **"i"** appears in the training data, it is followed by **"love"** with probability **1.0**.

---

### Why Do We Need This?

Bigram probabilities allow us to:

- Predict the next word
- Build simple autocomplete systems
- Generate text
- Understand word relationships in a corpus

The word with the highest probability is usually chosen as the predicted next word.

## Bigram Language Model

A Bigram Model predicts a word based on the previous word.

Example:

Sentence:
"I love NLP"

Bigrams:
- (I, love)
- (love, NLP)

---

Bigram Probability Table

````python
bigram_probabilities = {}

for (word1, word2), count in bigram_counts.items():

    probability = count / word_counts[word1]

    bigram_probabilities[(word1, word2)] = probability

print("Bigram Probabilities:\n")

for pair, prob in bigram_probabilities.items():
    print(f"{pair}: {round(prob, 3)}")
````

Creating Next Word Predictor

````python

def predict_next_word(word):

    candidates = []
    weights = []

    for (w1, w2), prob in bigram_probabilities.items():
        if w1 == word:
            candidates.append(w2)
            weights.append(prob)

    if not candidates:
        return "No prediction available"

    return random.choices(candidates, weights=weights)[0]
````

Test Predictions

````python
test_words = [
    "i",
    "love",
    "machine",
    "natural"
]

for word in test_words:

    prediction = predict_next_word(word)

    print(f"{word} → {prediction}")
````

Generate Text

````python
def generate_text(start_word, length=5):

    sentence = [start_word]

    current_word = start_word

    for _ in range(length):

        next_word = predict_next_word(current_word)

        if next_word == "No prediction available":
            break
        
        if len(sentence) > 1 and next_word == sentence[-1]:
            break

        sentence.append(next_word)

        current_word = next_word

    return " ".join(sentence)
````

Generate Example Sentences

````python
print("Generated text:", generate_text("i"))
print("Generated text:",generate_text("machine"))
print("Generated text:",generate_text("natural"))
````

````python
user_word = "deep"

print("Prediction:")
print(predict_next_word(user_word))
````

Checking multiple predictions

````python
word = "machine"

print(f"🔍 Multiple predictions for: {word}\n")

for i in range(5):
    print(f"Run {i+1}: {predict_next_word(word)}")
````

````python
test_words = ["hello", "xyz", "deep", "processing", "machine"]

print("🔍 Edge Case Testing:\n")

for word in test_words:
    print(f"{word} → {predict_next_word(word)}")
````

Checking bigram existence

````python
check_words = ["deep", "learning", "machine"]

print("🔍 Bigram existence check:\n")

for word in check_words:
    pairs = [pair for pair in bigram_probabilities if pair[0] == word]
    print(f"{word}: {pairs}")
````
