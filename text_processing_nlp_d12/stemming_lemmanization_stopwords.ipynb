# Stemming, Lemmatization, and Stopwords in NLP using NLTK

## Overview

This notebook introduces important NLP preprocessing techniques using NLTK:

- Stopwords Removal
- Stemming
- Lemmatization

These preprocessing steps help clean and normalize raw text before applying machine learning or NLP models.

---


# What are Stopwords?

Stopwords are common words that usually do not add important meaning to a sentence.

Examples:
- is
- the
- and
- in

Removing stopwords helps reduce unnecessary words from text data.

---


```python
# Import NLTK library
import nltk

# Import tokenizer
from nltk.tokenize import word_tokenize

# Import stopwords dataset
from nltk.corpus import stopwords

# Import stemming class
from nltk.stem import PorterStemmer

# Import lemmatization class
from nltk.stem import WordNetLemmatizer
```

Download Required NLTK Data

```python
# Download tokenizer data
nltk.download('punkt')

# Download stopwords dataset
nltk.download('stopwords')

# Download WordNet dataset for lemmatization
nltk.download('wordnet')
```

Create Sample Text

```python
# Sample sentence for NLP preprocessing
text = "NLTK is helping students learn preprocessing techniques in Natural Language Processing."

# Print original text
print("Original Text:")
print(text)
```

Tokenization

```python
# Convert text into individual words (tokens)
tokens = word_tokenize(text)

# Print tokens
print("Tokens:")
print(tokens)
```

Load Stopwords

```python
# Load English stopwords
stop_words = set(stopwords.words('english'))

# Print some stopwords
print("Stopwords Example:")
print(list(stop_words)[:10])
```

Remove Stopwords

```python
# Remove stopwords from token list
filtered_words = [word for word in tokens if word.lower() not in stop_words]

# Print filtered words
print("After Stopwords Removal:")
print(filtered_words)
```

# What is Stemming?

Stemming reduces words to their root form.

Examples:
- running → run
- studies → studi

Stemming is fast but may not always produce meaningful words.

---

```python
# Create Porter Stemmer object
stemmer = PorterStemmer()
```

```python
# Sample words for stemming
words = ["running", "runner", "runs", "easily", "studies", "happily", "tolerance"]

print("Stemming Results:\n")

# Apply stemming on each word
for word in words:
    stem = stemmer.stem(word)
    print(f"{word} -> {stem}")
```

```python
# Apply stemming on filtered words
stemmed_words = [stemmer.stem(word) for word in filtered_words]

# Print stemmed words
print("Stemmed Words:")
print(stemmed_words)
```

# What is Lemmatization?

Lemmatization converts words into meaningful base forms using vocabulary and grammar rules.

Examples:
- better → good
- flies → fly

Lemmatization is more accurate than stemming.

---

# Stemming vs Lemmatization

| Feature | Stemming | Lemmatization |
|--------|-----------|---------------|
| Output | Root form | Meaningful base word |
| Accuracy | Less accurate | More accurate |
| Speed | Faster | Slower |
| Example | studies → studi | studies → study |


```python
# Create WordNet Lemmatizer object
lemmatizer = WordNetLemmatizer()
```

```python
# Sample words for lemmatization
words = ["running", "runner", "runs", "better", "studies", "happily", "tolerance"]

print("Lemmatization Results:\n")

# Apply lemmatization on each word
for word in words:
    lemma = lemmatizer.lemmatize(word)
    print(f"{word} -> {lemma}")
```

```python
# Apply lemmatization on filtered words
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

# Print lemmatized words
print("Lemmatized Words:")
print(lemmatized_words)
```

```python
# Print all outputs together

print("Original Tokens:")
print(tokens)

print("\nFiltered Words:")
print(filtered_words)

print("\nStemmed Words:")
print(stemmed_words)

print("\nLemmatized Words:")
print(lemmatized_words)
```

```python
# Import spaCy library
import spacy
```

```python
# Load pre-trained English NLP model
nlp = spacy.load("en_core_web_sm")
```

```python
# Sample text for preprocessing
text = "NLTK is helping students learn preprocessing techniques in Natural Language Processing."

print("Original Text:")
print(text)
```

```python
# Process text using spaCy NLP pipeline
doc = nlp(text)

# Extract tokens in list
tokens = [token.text for token in doc]

print("Tokens:")
print(tokens)
```

```python
# Remove stopwords using spaCy built-in stopword list
filtered_words = [token.text for token in doc if not token.is_stop]

print("After Stopwords Removal:")
print(filtered_words)
```

```python
# Convert words to their base form (lemma)
lemmas = [token.lemma_ for token in doc]

print("Lemmatized Words:")
print(lemmas)
```

```python
# Sample text
text = "NLTK is AMAZING and helps students learn NLP preprocessing techniques easily."

print("Original Text:")
print(text)

# Step 1: Tokenization
tokens = word_tokenize(text)

print("\nTokens:")
print(tokens)

# Step 2: Stopwords Removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("\nAfter Stopwords Removal:")
print(filtered_tokens)

# Step 3: Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nStemmed Words:")
print(stemmed_words)

# Step 4: Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Words:")
print(lemmatized_words)
```

```python
# Sample text
text = "NLTK and spaCy are AMAZING tools for NLP preprocessing techniques."

print("Original Text:")
print(text)

# Step 1: Process text
doc = nlp(text)

# Step 2: Tokenization
tokens = [token.text for token in doc]

print("\nTokens:")
print(tokens)

# Step 3: Stopwords Removal
filtered_tokens = [token.text for token in doc if not token.is_stop]

print("\nAfter Stopwords Removal:")
print(filtered_tokens)

# Step 4: Lemmatization
lemmas = [token.lemma_ for token in doc]

print("\nLemmatized Words:")
print(lemmas)
```
