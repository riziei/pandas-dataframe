# Regular Expressions (Regex) in NLP

## Introduction

### What is Regex?

Regular Expressions (Regex) are patterns used to search, match, and manipulate text.

In NLP, regex is used for:

- Cleaning text  
- Removing unwanted characters  
- Extracting useful patterns  
- Preprocessing raw data  

## Basic Concepts
- Pattern matching
- Text searching and manipulation
- Role of Regex in text preprocessing

---
## Common Regex Patterns
- `\d` → digits
- `\d+` → numbers
- `\s` → whitespace
- `\s+` → multiple spaces
- `\w` → word characters
- `[A-Za-z]` → alphabets
- `[A-Za-z]+` → words
- `[^\w\s]` → punctuation removal

---
## NLP Applications of Regex
- Text cleaning
- Removing punctuation
- Removing numbers
- Removing extra spaces
- Extracting words
- Extracting patterns (emails, dates, hashtags)

---
## Regex Functions in Python
- `re.sub()` → replace patterns
- `re.findall()` → extract patterns
- `re.search()` → find first match
- `re.match()` → match at beginning

---

## Regex in NLP Pipeline
- Lowercasing text
- Cleaning noise (punctuation, numbers)
- Normalizing whitespace
- Preparing text for tokenization

---

## Key Takeaways
- Regex is essential for text preprocessing
- Helps convert raw text into clean structured data
- Frequently used before tokenization and modeling

---

```python
# Import regular expression module
import re
```

```python
# Sample text
text = "NLTK!!! is amazing... NLP 2025 is    evolving fast!!!"

print("Original Text:")
print(text)
```

Remove punctuation

```python
# Remove all punctuation using regex
clean_text = re.sub(r'[^\w\s]', '', text)

print("\nAfter Removing Punctuation:")
print(clean_text)
```

Remove numbers

```python
# Remove digits
no_numbers = re.sub(r'\d+', '', clean_text)

print("\nAfter Removing Numbers:")
print(no_numbers)
```

Remove extra space

```python
# Replace multiple spaces with single space
clean_spaces = re.sub(r'\s+', ' ', no_numbers).strip()

print("\nAfter Removing Extra Spaces:")
print(clean_spaces)
```

```python
# Extract only words (A-Z, a-z)
words = re.findall(r'[A-Za-z]+', text)

print("\nExtracted Words:")
print(words)
```

```python
# Sample text containing emails, phone numbers, and URLs
text = """
Hello team,

Contact us at support@gmail.com or admin123@company.org.
You can also reach us at +977-9812345678 or 9801234567.

Visit our website:
https://www.example.com
http://nlp-project.org/docs

Thank you!
"""

print("Original Text:\n")
print(text)
```

```python
# Email pattern (Gmail + general emails)
# \S+  → matches non-space characters
# @    → matches @ symbol
# \S+  → matches domain part (like gmail.com)
email_pattern = r'\S+@\S+'

# Phone number pattern (simple international + local formats)
# \+?\d{1,3}? → optional country code (+977)
# [-\s]?      → optional separator (- or space)
# \d{7,10}    → main phone number digits
phone_pattern = r'\+?\d{1,3}?[-\s]?\d{7,10}'

# URL pattern (http/https websites)
# https?://   → matches http or https
# \S+         → matches all non-space characters (rest of the URL)
url_pattern = r'https?://\S+'
```

```python
#Extract emali
emails = re.findall(email_pattern, text)
print("\nExtracted Emails:")
print(emails)

#Extract phone number
phones = re.findall(phone_pattern, text)
print("\nExtracted Phone Numbers:")
print(phones)

#Extract urls
urls = re.findall(url_pattern, text)
print("\nExtracted URLs:")
print(urls)
```
