# Bag of Words (BoW), TF-IDF, and Document Similarity

## Overview

This notebook introduces two fundamental text representation techniques used in Natural Language Processing (NLP):

- Bag of Words (BoW)
- Term Frequency-Inverse Document Frequency (TF-IDF)

These techniques convert text into numerical representations that machine learning algorithms can process and analyze.

The notebook also demonstrates how to measure similarity between documents using TF-IDF vectors and Cosine Similarity.

---

## What is Bag of Words (BoW)?

Bag of Words (BoW) is a text representation technique that converts text into numerical vectors based on word frequency.

Key characteristics:

- Creates a vocabulary of unique words
- Counts how many times each word appears
- Ignores grammar and word order
- Represents documents as numerical vectors

### Example

Document 1:
```
I love NLP
```

Document 2:
```
I love Python
```

Vocabulary:
```
[I, love, NLP, Python]
```

Vector Representation:

| Document | I | love | NLP | Python |
|-----------|---|------|-----|--------|
| Doc 1 | 1 | 1 | 1 | 0 |
| Doc 2 | 1 | 1 | 0 | 1 |

---

## Advantages of BoW

- Simple and easy to understand
- Fast to implement
- Works well for basic text classification tasks

---

## Limitations of BoW

- Ignores word order
- Does not capture meaning
- Treats all words equally
- Common words may dominate the representation

---

````python
# Import required libraries
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
# Sample text documents
documents = [
    "I love natural language processing",
    "Natural language processing is amazing",
    "I enjoy learning machine learning",
    "Machine learning is part of artificial intelligence"
]

# Print documents
for i, doc in enumerate(documents, 1):
    print(f"Document {i}: {doc}")
````

Bag of Words (CountVectorizer)

````python
# Create BoW model
bow_vectorizer = CountVectorizer()

# Fit and transform documents
bow_matrix = bow_vectorizer.fit_transform(documents)

# Convert to dataframe for better visualization
bow_df = pd.DataFrame(
    bow_matrix.toarray(),
    columns=bow_vectorizer.get_feature_names_out()
)

print("Bag of Words Matrix:")
bow_df
````

## What is TF-IDF?

TF-IDF stands for:

- TF = Term Frequency
- IDF = Inverse Document Frequency


### 1. Term Frequency (TF)
Measures how often a word appears in a document.

Example:
If a word appears many times → TF is high

---

### 2. Inverse Document Frequency (IDF)
Measures how important a word is across all documents.

- Common words (e.g., "is", "the") → low importance  
- Rare words (e.g., "NLP") → high importance  


TF-IDF improves upon BoW by assigning importance scores to words.

Words that appear frequently in a document but rarely across other documents receive higher weights.

---

## Why TF-IDF?

Not all words are equally important.

For example:

```
the, is, and
```

appear in many documents and provide little useful information.

TF-IDF reduces the importance of such common words and highlights more meaningful terms.

---

## Advantages of TF-IDF

- Considers word importance
- Reduces influence of common words
- Produces better document representations
- Widely used in information retrieval systems

---

TF-IDF Vectorization

````python
# Create TF-IDF model
tfidf_vectorizer = TfidfVectorizer()

# Transform documents
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Convert to dataframe
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

print("TF-IDF Matrix:")
tfidf_df
````

## Sparsity Problem

BoW and TF-IDF create large matrices where:
- Most values are 0

This is called a sparse matrix.

Why it happens:
- Large vocabulary size
- Many words do not appear in every document

Impact:
- High memory usage
- Slower processing (in large datasets)

---

Sparsity Example

````python
import pandas as pd

# Convert BoW matrix to dataframe
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

print("BoW Matrix (Sparse Representation):")
print(df)

# Count zeros
zero_count = (df == 0).sum().sum()

print("\nNumber of zero values (sparsity):", zero_count)
````

#### Key Evidence from our Matrices:
* Unique Word Boost: In Document 0, two words both have a BoW count of `1`. However, the word unique only to Document 0 gets a higher TF-IDF score (`0.59`) than the word shared with Document 1 (`0.46`).
* Length Normalization: TF-IDF automatically scales scores based on document length, preventing longer documents from artificially dominating the feature space.
* Peak Importance: The highest score in the dataset (`0.77` in Document 2) occurs where a word has both a high repetition within that document *and* high scarcity across the collection.

## Bag of Words vs TF-IDF

| Feature | BoW | TF-IDF |
|----------|----------|----------|
| Uses Word Counts | Yes | No |
| Considers Importance | No | Yes |
| Handles Common Words Better | No | Yes |
| Information Quality | Basic | Better |
| Practical Usage | Simple Tasks | Real-world NLP |
