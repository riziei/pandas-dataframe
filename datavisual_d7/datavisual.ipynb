Importing libraries

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

<h2>Load dataset</h2>
The Iris dataset contains flower measurements:Sepal length,Sepal width,Petal length,Petal width
<br>
Target represents flower species.

```python
# Load iris dataset
iris = load_iris()

# Convert to pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add target column
df['target'] = iris.target

# Show first 5 rows
df.head()
```

Eplore dataset

```python
#print(df.shape)
print(df.info())
#print(df.describe())
print(df['target'].value_counts())
```

```python
df['species'] = df['target'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})
```

```python
# Count of each species
df["species"].value_counts().plot(kind="bar")

plt.title("Count of Flower Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

<h1>Histograms</h1>
Histogram shows how values are distributed across ranges.

```python
plt.hist(df["sepal length (cm)"])

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["sepal width (cm)"])

plt.title("Distribution of Sepal width")
plt.xlabel("Sepal width")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["petal length (cm)"])

plt.title("Distribution of petal Length")
plt.xlabel("petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["petal width (cm)"])

plt.title("Distribution of petal width")
plt.xlabel("petal width")
plt.ylabel("Frequency")
plt.show()
```

<h1>Scatter plot</h1>
Scatter plot shows the relationship between two variables.

```python
plt.scatter(df["sepal length (cm)"], df["petal length (cm)"])

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
plt.scatter(df["sepal width (cm)"], df["petal width (cm)"])

plt.title("Sepal width vs Petal width")
plt.xlabel("Sepal width")
plt.ylabel("Petal width")
plt.show()
```

```python
#adding color variation
colors = {'setosa':'red', 'versicolor':'blue', 'virginica':'green'}

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['petal length (cm)'], color=colors[species], label=species)
plt.title("Sepal length vs Petal length (by species)")

plt.legend()
plt.show()
```

```python
colors = {'setosa':'red', 'versicolor':'blue', 'virginica':'green'}

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal width (cm)'], subset['petal width (cm)'],color=colors[species], label=species)
plt.title("Sepal Width vs Petal Width (by species)")

plt.legend()
plt.show()
```
