## Objective

Today's goal was to understand how words can be represented as numerical vectors using Word Embeddings and to implement a Word2Vec model using Gensim.

The focus was on learning how machines capture semantic relationships between words beyond simple frequency counts.

---


## What are Word Embeddings?

Word Embeddings are dense vector representations of words where words with similar meanings have similar vector representations.

Unlike one-hot encoding, embeddings capture semantic relationships between words.

Example:

* King  → [0.25, -0.13, 0.78, ...]
* Queen → [0.22, -0.10, 0.75, ...]
* Man   → [0.15, -0.40, 0.50, ...]
* Woman → [0.12, -0.35, 0.48, ...]

---

## Limitations of One-Hot Encoding

One-hot encoding creates sparse vectors:

* Dog  → [1, 0, 0, 0]
* Cat  → [0, 1, 0, 0]

Problems:

- Very large vectors
- No semantic meaning
- Dog and Cat appear completely unrelated

---


## What is Word2Vec?

Word2Vec is a neural network-based algorithm that learns word embeddings from text.

The model learns word relationships by observing surrounding context words.

Main idea:

Words appearing in similar contexts
tend to have similar meanings.

Example:

- The cat sat on the mat.
- The dog sat on the floor.

Word2Vec learns that:

cat ≈ dog

because both appear in similar contexts.

---

## 4. Word2Vec Architectures

### CBOW (Continuous Bag of Words)

Predicts the target word from surrounding context words.

Example:

Context:
"The cat ___ milk"

Prediction:
"drinks"

Advantages:

- Faster training
- Works well on smaller datasets

---

### Skip-Gram

Predicts surrounding words from a target word.

Example:

Target:
"drinks"

Predict:

- cat
- milk

Advantages:

- Better for rare words
- Usually produces higher-quality embeddings

---


## Word2Vec Architecture

Word2Vec is a shallow neural network that learns embeddings by processing large amounts of text.

Main architectures:

- CBOW
- Skip-Gram

Output:

Word → Dense Vector Representation

---

## Gensim Library

Gensim provides an easy implementation of Word2Vec.

It handles:

- Training
- Vocabulary building
- Similarity calculations
- Vector retrieval

Install and Import Libraries

```python
# Install Gensim 
#pip install gensim

# Import required libraries
import nltk
from gensim.models import Word2Vec
from gensim.test.utils import common_texts
```

Create Sample Dataset

```python
sentences = [
    "machine learning is amazing",
    "deep learning is subset of machine learning",
    "artificial intelligence includes machine learning",
    "neural networks are used in deep learning",
    "data science uses machine learning",
    "deep learning requires lot of data",
    "nlp is a field of artificial intelligence"
]
```

Text Preprocessing

```python
from nltk.tokenize import word_tokenize

sentences = [word_tokenize(s.lower()) for s in sentences]

print("Tokenized Sentences Used for Training:\n")

for i, sentence in enumerate(sentences, start=1):
    print(f"Sentence {i}: {sentence}")
```

Train a Word2Vec Model (Skip-Gram Default)

```python
# Train Word2Vec model

model = Word2Vec(
    sentences,
    vector_size=100,   # Dimension of word vectors
    window=3,          # Context window size
    min_count=1,       # Include all words
    workers=4,         # CPU cores
    sg=1               # 1 = Skip-Gram, 0 = CBOW
)

print("Model trained successfully!")
```

View Vocabulary

```python
# Display all learned words

print("Vocabulary:")

for word in model.wv.index_to_key:
    print(word)
```

Check Vector for a Word

```python
# Get vector representation of a word

vector = model.wv["machine"]

print("Vector for 'machine':")
print(vector)

print("\nVector Length:", len(vector))
```

Find Similar Words

```python
# Find words similar to "machine"

similar_words = model.wv.most_similar("machine")

print("Words similar to 'machine':")

for word, score in similar_words:
    print(word, round(score, 3))
```

Calculate Similarity Between Two Words

```python
# Similarity score between words

score = model.wv.similarity(
    "machine",
    "learning"
)

print("Similarity Score:")
print(score)
```

Compare Multiple Words

```python
words = [
    ("machine", "learning"),
    ("machine", "data"),
    ("machine", "networks")
]

for w1, w2 in words:

    score = model.wv.similarity(w1, w2)

    print(f"{w1} <-> {w2}: {score:.3f}")
```

Word Analogy

```python
# Vector arithmetic

result = model.wv.most_similar(
    positive=["deep", "machine"],
    negative=["data"],
    topn=5
)

print("Analogy Results:")

for word, score in result:
    print(word, round(score, 3))
```

Show Top Similar Words for Every Word

```python
for word in model.wv.index_to_key:

    print("\nWord:", word)

    similar = model.wv.most_similar(word, topn=3)

    for sim_word, score in similar:
        print("  ", sim_word, round(score, 3))
```

Visual Understanding

```python
test_words = [
    "machine",
    "learning",
    "deep"
]

for word in test_words:

    print("\nWord:", word)

    print("Most Similar:")

    for similar_word, score in model.wv.most_similar(word, topn=5):

        print(
            f"{similar_word:<15} {score:.3f}"
        )
```

Explore Vocabulary Size

```python
print("Vocabulary Size:")
print(len(model.wv))
```

Save the Model

```python
# Save trained model

model.save("word2vec.model")

print("Model saved.")
```

Show First 10 Dimensions of Each Word Vector

CBOW vs Skip-Gram Comparison

```python
# CBOW MODEL (sg = 0)
cbow_model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    workers=4,
    sg=0   # CBOW
)

print("CBOW model trained")


# SKIP-GRAM MODEL (sg = 1)
skipgram_model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    workers=4,
    sg=1   # Skip-Gram
)

print("Skip-Gram model trained")

print("\nCBOW vs Skip-Gram Similar Words Comparison\n")

cbow_sim = cbow_model.wv.most_similar("machine", topn=5)
skip_sim = skipgram_model.wv.most_similar("machine", topn=5)

print("CBOW (sg=0) - Most similar to 'machine':")
for word, score in cbow_sim:
    print(f"  {word:<15} score: {score:.4f}")

print("\nSkip-Gram (sg=1) - Most similar to 'machine':")
for word, score in skip_sim:
    print(f"  {word:<15} score: {score:.4f}")
```
