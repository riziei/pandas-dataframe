## What is Tokenization?

Tokenization is the process of splitting text into:
- words
- sentences

---

Load NLTK Model

```python
import nltk

# Download the modern tokenizer package
nltk.download('punkt_tab')
```

Word Tokenization using NLTK

```python
# Import word tokenizer
from nltk.tokenize import word_tokenize

# Sample text
text = "AI is transforming education."

# Convert text into tokens
tokens = word_tokenize(text)

# Display tokens
print(tokens)
```

Sentence Tokenization using NLTK

```python
# Import sentence tokenizer
from nltk.tokenize import sent_tokenize

# Sample paragraph
text = "AI is amazing. NLP is interesting."

# Split paragraph into sentences
sentences = sent_tokenize(text)

# Display result
print(sentences)
```

# Features of spaCy

spaCy provides:
- tokenization
- POS tagging
- Named Entity Recognition (NER)
- dependency parsing
- lemmatization
- sentence segmentation

---

Load spaCy Model

```python
import spacy

# Load English NLP model
nlp = spacy.load("en_core_web_sm")
```

Word Tokenization

```python
# Sample text
text = "AI is transforming education."

# Process text using spaCy
doc = nlp(text)

# Print tokens
for token in doc:
    print(token.text)
```

Sentence Tokenization

```python
# Sample paragraph
text = "AI is amazing. NLP is interesting."

# Process text
doc = nlp(text)

# Print sentences
for sentence in doc.sents:
    print(sentence)
```

Comparing NLTK and spaCy

```python
# Sample text
text = "Natural Language Processing is fun!"

print('\n Original text')
print(text)

print("\nNLTK Tokens:")
print(word_tokenize(text))


print("\nspaCy Tokens:")

# Process text using spaCy
doc = nlp(text)

# Display tokens
for token in doc:
    print(token.text)
```

```python
#Multiple sentences
texts = [
    "I love Natural Language Processing!",
    "NLTK vs spaCy: which one is better?",
    "Tokenization is the first step in NLP."
]
print('\n Multiple sentences')
print(texts)


print(" NLTK Tokenization ")
for text in texts:
    print("\nText:", text)
    print("Words:", word_tokenize(text))

print("\n spaCy Tokenization ")
for text in texts:
    doc = nlp(text)
    print("\nText:", text)
    print("Tokens:", [token.text for token in doc])
    
```

```python
#punctuation sentence
text = "Hello!!! Are you learning NLP??? Yes, I am."

print("\n Punctuation-heavy Text ")
print(text)

print("\nNLTK:", word_tokenize(text))

doc = nlp(text)
print("spaCy:", [token.text for token in doc])
```

```python
#paragraph tokenization
paragraph = """
Natural Language Processing is a field of Artificial Intelligence.
It helps computers understand human language.
NLTK and spaCy are popular libraries for NLP tasks.
"""

paragraph = paragraph.strip()#\n in op remove
print("\n Long Paragraph ")
print(paragraph)

print("\nNLTK Sentence Tokenization:")
print(sent_tokenize(paragraph))

print("\nspaCy Sentence Tokenization:")
doc = nlp(paragraph)
for sent in doc.sents:
    print(sent.text)
```

```python
paragraph="""
Hello!!!!Welcome to NLP world!!!
Are you excited!!!! I hope so!!
NTK is simple...... spaCy is powerful!!!!
Lets learn nlp step-by-step!!!!!
"""
paragraph = paragraph.strip()
print('\n original paragraph')
print(paragraph)

print('\n NLTK word tokenization')
nltk_tokens=word_tokenize(paragraph)
print(nltk_tokens)

print("\n NLTK sentence tokenization")
nltk_sentences=sent_tokenize(paragraph)
print(nltk_sentences)

print('\n spaCy tokenization')
doc=nlp(paragraph)

spacy_tokens=[token.text for token in doc]
print('Tokens: ',spacy_tokens)

print('\n spaCy sentence tokenization')
for sent in doc.sents:
    print(sent.text)
```

# Conclusion: Tokenization (NLTK vs spaCy)

## Summary

Tokenization is the first and most important step in Natural Language Processing (NLP). It involves breaking raw text into smaller units called tokens, such as words, punctuation marks, or sentences.

In this notebook, we explored tokenization using two popular NLP libraries:
- NLTK (Natural Language Toolkit)
- spaCy

---

## Key Learnings

- Tokenization converts text into structured units for further processing.
- NLTK provides simple and easy-to-use functions for tokenization.
- spaCy provides faster and more advanced tokenization using an NLP pipeline.
- Both libraries handle punctuation by splitting it into separate tokens.
- Output may look similar for basic text, but spaCy offers richer internal token information.

---

## NLTK vs spaCy (Tokenization)

| Feature | NLTK | spaCy |
|--------|------|-------|
| Output Type | List of strings | Token objects |
| Ease of Use | Simple | Structured |
| Speed | Slower | Faster |
| Best Use | Learning & research | Production NLP |
| Extra Information | Limited | POS, lemma, dependency available |

---

## Final Insight

Tokenization is a foundational step in NLP that prepares text for advanced tasks like:
- Part-of-speech tagging
- Named entity recognition
- Sentiment analysis
- Text classification

NLTK is ideal for understanding the basics, while spaCy is better suited for real-world applications due to its speed and advanced features.

---

## Conclusion

Both NLTK and spaCy are powerful tools for tokenization. The choice between them depends on the use case: NLTK for learning and experimentation, and spaCy for efficient and production-level NLP pipelines.
