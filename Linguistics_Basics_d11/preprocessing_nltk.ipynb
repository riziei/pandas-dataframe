Basic NLP Preprocessing Code Examples

```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = "NLTK is AMAZING!!! It helps with NLP preprocessing tasks like tokenization, stemming, and lemmatization."
print("Original Text:\n", text)
```

Lowercasing


```python
text = text.lower()
print("\nLowercased Text:\n", text)
```

Remove Punctuation

```python
text = text.translate(str.maketrans('', '', string.punctuation))
print("\nWithout Punctuation:\n", text)
```

Tokenization

```python
tokens = word_tokenize(text)
print("\nTokens:\n", tokens)
```

Remove Stopwords

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]

print("\nAfter Stopwords Removal:\n", filtered_tokens)
```

Stemming

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nStemmed Words:\n", stemmed_words)
```

Lemmatization

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Words:\n", lemmatized_words)
```

Final NLP Preprocessing Pipeline

```python
text = "NLTK is AMAZING!!! It helps with NLP preprocessing tasks like tokenization, stemming, and lemmatization."

# 1. Lowercase
text = text.lower()

# 2. Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# 3. Tokenize
tokens = word_tokenize(text)

# 4. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w not in stop_words]

# 5. Stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered]

# 6. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]

print("\nFinal Tokens:", tokens)
print("\nAfter Stopwords:", filtered)
print("\nStemmed:", stemmed)
print("\nLemmatized:", lemmatized)
```
